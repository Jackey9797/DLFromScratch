```python
%load_ext autoreload
%autoreload 2
import tc

class MLP:
    def __init__(self, input_size, output_size, hidden_size, layer_num, activation) -> None:
        if activation == 'R': 
            self.activation = tc.ReLU() 
        elif activation == 'S': 
            self.activation = tc.Sigmoid() 

        self.layer_list = [tc.Linear(input_size, hidden_size), self.activation] 
        
        for i in range(layer_num): 
            self.layer_list += [tc.Linear(hidden_size, hidden_size), self.activation]    

        self.layer_list += [tc.Linear(hidden_size, output_size)]
        self.net = tc.Sequential(self.layer_list)

    def __call__(self, x):
        y = self.net(x) 
        return y
```
